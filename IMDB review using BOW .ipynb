# Checking if this contains any html tags

```python
import re
for i in df['review'].values:
    if(len(re.findall('<.*?>', i))):
        print(i)
        break
```

```python
def cleanhtml(sentence):
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, ' ', sentence)
    return cleantext
def cleanpunc(sentence): #function to clean the word of any punctuation or special characters
    cleaned = re.sub(r'[?|!|\'|"|#]',r'',sentence)
    cleaned = re.sub(r'[.|,|)|(|\|/]',r' ',cleaned)
    return  cleaned
```

# Loading the stop words

```python
from nltk.corpus import stopwords
stop = set(stopwords.words('english'))
print(stop)
```

# Loading the stemmer

```python
from nltk.stem import SnowballStemmer
sno = SnowballStemmer('english')
```

# Cleaning and preprocessing the data

```python
str1 = ' '
#fil_words = []
str = ' '
final_string = []

for i in df['review'].values:
    fil_word = []
    sent = cleanhtml(i)
    for w in sent.split():
        for cleanedwords in cleanpunc(w).split():
            if((cleanedwords.isalpha()) and (len(cleanedwords) > 2)):
                if(cleanedwords.lower() not in stop):
                    s = (sno.stem((cleanedwords.lower())).encode('utf8'))
                    fil_word.append(s)
    
    str1 = b' '.join(fil_word)
    final_string.append(str1) 
                    
```

# Creating a colum in data frame of preprocessed data

```python
df['cleanedtext'] = final_string
```

# Initializing bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(2, 4)) # 2-grams, 3-grams and 4 -grams
```

# separating the data into features and labels

```python
x = df['cleanedtext']
y = df['label']
```

# Applying bag of words

```python
x = cv.fit_transform(x)
```

```python
x.get_shape()
```

# Spliting the data into train and test

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=0)
```

# Appyling and checking accuracy of Multinomial Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB().fit(x_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
pred = classifier.predict(x_test)
print(classification_report(y_test, pred))
print()
print('Confusion Matrix:\n',confusion_matrix(y_test, pred))
print()
print('Accuracy : ',accuracy_score(y_test, pred))
```

# Appyling and checking accuracy of Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(x_train, y_train)
```

```python
pred = clf.predict(x_test)
print(classification_report(y_test, pred))
print()
print('Confusion Matrix:\n',confusion_matrix(y_test, pred))
print()
print('Accuracy : ',accuracy_score(y_test, pred))
```

#  Appyling and checking accuracy of Decision Tree Classifier

```python
from sklearn import tree
clf_tree = tree.DecisionTreeClassifier()
clf_tree.fit(x_train,y_train)
```

```python
pred = clf_tree.predict(x_test)
print(classification_report(y_test, pred))
print()
print('Confusion Matrix:\n',confusion_matrix(y_test, pred))
print()
print('Accuracy : ',accuracy_score(y_test, pred))
```
